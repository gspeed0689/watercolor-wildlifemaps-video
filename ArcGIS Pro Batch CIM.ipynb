You can access layers with the Cartographic Information Model (CIM) and make bulk changes in ArcPy. 

```python
import arcpy
```

Get the layering map.

```python
project = arcpy.mp.ArcGISProject('current')
maps = project.listMaps()
layering_map = [x for x in maps if x.name == "Layering"][0]
print(layering_map.name)
```

```python
layers = layering_map.listLayers()
```

This method uses the [CIM](https://pro.arcgis.com/en/pro-app/3.1/arcpy/mapping/python-cim-access.htm) to batch change the stretching of the rasters. 

```python
for layer in layers:
    if layer.isRasterLayer:
        print(layer.dataSource)
        cim = layer.getDefinition('V3')
        cim.colorizer.stretchType = "MinimumMaximum"
        cim.colorizer.stretchStatsType = "Dataset"
        cim.colorizer.useCustomStretchMinMax = True
        cim.colorizer.customStretchMaxBlue = 255
        cim.colorizer.customStretchMaxGreen = 255
        cim.colorizer.customStretchMaxRed = 255
        layer.setDefinition(cim)
```
